```python

tsvin = open("flightData/flight_edges.tsv","r")
tsvin = csv.reader(tsvin, delimiter='\t')

before = []
during = []
after = []

totalBefore = []
totalDuring = []
totalAfter = []

for row in tsvin:
    if row[0] == "MSY" and row[8] == "200507":
        before.append(int(row[4]))
    if row[0] == "MSY" and row[8] == "200508":
        during.append(int(row[4]))
    if row[0] == "MSY" and row[8] == "200509":
        after.append(int(row[4]))
    if row[0] == "MSY" and row[8][4:] == "07":
        totalBefore.append(int(row[4]))
    if row[0] == "MSY" and row[8][4:] == "08":
        totalDuring.append(int(row[4]))
    if row[0] == "MSY" and row[8][4:] == "09":
        totalAfter.append(int(row[4]))
```

```python
plt.plot([sum(before),sum(during),sum(after)],label='In 2006')
plt.plot([sum(totalBefore)/20.0,sum(totalDuring)/20.0,sum(totalAfter)/20.0],label='Average 1990-2009')
plt.xticks( arange(3), calendar.month_name[8:11])
plt.title("Trends out of New Orleans (MSY) around Katrina")
plt.ylabel("Passengers Flying")
plt.xlabel("Month")
plt.legend()
```
